# Oodle Customer Segmentation

**Problem Statement**

Oodle wants to identify their customer segments that need more marketing investments to further grow their business through more incisive targeted marketing initiatives. The dataset provided has demographic information of 10,000 records, and has data about the loan application. 

**Metadata of dataset**:
1. Age : Age of the applicant
2. Car_type : Type of car the loan application was for, one of Convertible, Saloon and SUV
3. Loanamount : Loan amount applied for
4. Deposit : Customer deposit
5. Area : Based on the postcode of the applicant whether they live in an urban or rural area
6. APR : APR on the loan
7. Application_outcome: Whether the loan was approved or declined
8. Converted: Whether the application approval converted into a funded 

Through financing principles, we understand that APR of the loan primarily depends on the credit score of the loan applicant. Traditionally, credit scores are affected by socio-economic indicators. In our current case and dataset, the most prominent socio-economic indicator is "age" of the applicant. The APR of the loan can be considered as a proxy for the credit score of the applicant. 

Financial Institutions generally are able to generate high revenues from loan applicants who have a low % of deposit, and not very high credit scores. Hence, they can charge higher APRs for the approved loan. This follows the principle of high risk vs reward ratios. Oodle should strike the balance between risk and reward. If the organiation chooses customers who have high APR (low credit scores), the organization could potentialy make lots of money over time, but run a high risk of loan defaults. 
On the other hand, if the organization chooses customers who have low APR (high credit score), the value of servicing the customer is low, and the Customer Lifetime Value (CLV) is low. 

Hence, this exercise will identify customers using the following factors:
- Relatively higher APR
- Lower % of deposit
- Age

The two factors mentioned above are fairly indicative of the credit risk and customer value.

# Import packages

```python

import pandas as pd
import re
import numpy as np
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation
%matplotlib inline 
sns.set(color_codes=True)
from sklearn.cluster import KMeans
```

# Data Loading

```python
#Load data and rearrange columns

input=pd.read_csv('Case_Study_data_for_test.csv',sep=',')
input = input[['ID', 'age', 'car_type', ' Loanamount ', ' Deposit ', 'area', 'APR',
       'application_outcome', 'Funded ']]

#Renaming and arranging columns
input.columns=['ID', 'age', 'car_type', 'Loanamount', 'Deposit', 'area', 'APR',
       'application_outcome', 'Funded']
temp=input

```

# Data Preprocessing

```python
#Drop duplicates
temp= temp.drop_duplicates(['age', 'car_type', 'Loanamount', 'Deposit', 'area', 'APR',
       'application_outcome', 'Funded']).reset_index()
temp=temp[['ID', 'age', 'car_type', 'Loanamount', 'Deposit', 'area',
       'APR', 'application_outcome', 'Funded']]
input=temp

#Loan Amount and desposit have symbos and needs to be converted to floar
input['Loanamount']=(input['Loanamount'].replace( '[\£,)]','', regex=True ).astype(float))
input['Deposit']=(input['Deposit'].replace( '[\£,)]','', regex=True ).astype(float))

#Replace Nan values in APR with '0'
input['APR'] = input['APR'].fillna(0)
input['APR'] = input['APR']*100

#Calculate two columns that represent present value of Car and % of deposit
input['Present Value of Car']=input['Loanamount']+input['Deposit']
input['% of deposit']=input['Deposit']/input['Present Value of Car']*100
input['age_new'] = pd.cut(x=input['age'], bins=[17, 23, 29, 39, 49, 59, 65], labels=['18-23', '24-29', '30-39','40-49','50-59', '60-65'])
```

*Summary*
- Loan Amount and Deposit have the currency values (str), and need to be converted to float data type. 
- APR has NA values for declined loans. So we replace 'NA' with '0'. APR is represented as percentage.
- We introduce two additional columns- Present Value of Car, % of Deposit
- As of 2019, the average FICO score increases by age. For our case, age of the loan applicants are bucketed in the same way mentioned in the link provided. (Source:https://www.experian.com/blogs/ask-experian/research/credit-scores-by-age/#:~:text=The%20average%20FICO%C2%AE%20Score%20among%2050%2D%20to%2059%2Dyear,among%20those%20in%20their%2060s.)

# K-Means Clustering

**The cluster input variables are derived from the correlation matrix that depict low correlation coefficients. Moreover, from a car financing perspective, the important facets of a customer that depict credit worthiness and CLV differentiaition amongst customer segments are age, APR, and % of deposit.**

An important distinction for the clustering process is to filter the cluster input to only those loan applications that are approved. The model makes an assumption that the factors contributing to the rejection of the application is not contained within this dataset. Hence, we use the approved loan applications to identify customer segments that would benefit from targeted marketing initiatives.


# Test for Multi-collinearity

An important problem to circumvent before the clustering process is to eliminate the Garbage-In-Garbage-Out (GIGO) problem. In theory, the multi-collinearity problem could impact cluster integrity. 

Hence, we need to eliminate variables that display high correlation coefficient. Generally, if the correlation coefficient is greater than 0.85 on a scale of -1 to 1, we can ignore the variables for the clustering analysis. EDA has indicated that age, APR, and % of deposit impact outcomes of loan applications. 

```python
#Filter Approved Loan applications
cluster_input=input.loc[input['application_outcome']=='approved'].reset_index()
cluster_input=cluster_input[[ 'ID','age','APR', '% of deposit']]

#Generate correlation matrix
cor_mat= cluster_input.corr()
sns.heatmap(cor_mat*100, annot=True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(cluster_input)
data_scaled2=data_scaled
data_scaleddf=pd.DataFrame(data_scaled).describe()
```

One of the challenges of the K-means clustering technique is to identify the number of clusters, exposing the model to a marigin of error. Inertia can be defined as the measure of how internally coherent the clusters are. We need to try different samples (n=number of clusters) to minimize the inertia criterion. 

**Optimal number of Clusters**

```python
SSE = []
for cluster in range(1,20):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster, init='k-means++',random_state=3000)
    kmeans.fit(data_scaled)
    SSE.append(kmeans.inertia_)

# converting the results into a dataframe and plotting them
frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
frame2=frame
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```


The standard practice is to identify the cluster point at which the inertia levels start stabilizing. For our case, we estimate that the inertia levels start stabilizing when the number of clusters is **'10'**. This indicates that the clusters are most differentiated when the number of cluster is **'10'**. Hence, we will fit the K means cluster to n_cluster =10, and evaluate the results of the cluster. 


**Clusters= 10**

```python
kmeans = KMeans(n_jobs = -1, n_clusters =10, init='k-means++', random_state=3000)
kmeans.fit(data_scaled)
pred = kmeans.predict(data_scaled)
frame = pd.DataFrame(data_scaled)
frame['cluster'] = pred
clusters=pd.merge(cluster_input, frame, left_index=True, right_index=True)
clusters=clusters[['ID','cluster']]
new2=pd.merge(input, clusters, on='ID')
new2=new2[['ID',  'age_new','car_type',  'area', 'APR', 'application_outcome', 'Funded', 'Present Value of Car' , '% of deposit',  'cluster']]

```
